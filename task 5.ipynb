```python
import pandas as pd
```

```python
data =pd.read_csv('tested.csv')
```

```python
data.info()
```

```python
data.describe()
```

```python
print(data['Survived'].value_counts())
print(data['Pclass'].value_counts())
print(data['Sex'].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.pairplot(data, hue='Survived')
plt.show()
```

```python
numeric_cols = data.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = data[numeric_cols].corr()
plt.figure(figsize=(10,7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Age', hue='Survived', multiple='stack', bins=30)
plt.title('Age Distribution of Survivors vs Non-Survivors')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()



#Histograms reveal that the majority of passengers were young, with a significant drop-off in older age groups.
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Survived', y='Age', data=data)
plt.title('Boxplot of Age by Survival')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=data, alpha=0.6)
plt.title('Scatterplot of Age vs. Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
